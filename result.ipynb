```python
def process_campos_obrigatorios(df, columns):
    for column in columns:
        if column in df.columns:
            df = df.dropna(subset=[column])
            df = df[df[column].isin([1, 2])]
    return df
```

```python
def alrm_condicional(df):
    # Condi√ß√µes para os campos obrigat√≥rios
    mandatory_conditions = (df['CLASSI_FIN'].isin([11, 12]) & df['CRITERIO'].isin([1, 2]))

    # Lista de campos a serem verificados
    fields = ['ALRM_SANG', 'ALRM_HEMAT', 'ALRM_HEPAT', 'ALRM_LIQ', 'ALRM_LETAR', 'ALRM_ABDOM', 'ALRM_VOM', 'ALRM_PLAQ', 'ALRM_HIPOT', 'DT_ALRM']

    for field in fields:
        # Remover linhas onde o campo √© obrigat√≥rio mas est√° nulo
        df = df[~(mandatory_conditions & df[field].isna())]
        # Remover linhas onde o campo est√° preenchido mas a condi√ß√£o n√£o √© satisfeita
        df = df[~((~mandatory_conditions) & df[field].notna())]

    return df

```

```python
def grav_condicional(df):
    # Condi√ß√µes para os campos obrigat√≥rios
    mandatory_conditions = (df['CLASSI_FIN'] == 12) & (df['CRITERIO'].isin([1, 2]))

    # Lista de campos a serem verificados
    fields = [
        'GRAV_PULSO', 'GRAV_CONV', 'GRAV_ENCH', 'GRAV_INSUF', 'GRAV_TAQUI', 
        'GRAV_EXTRE', 'GRAV_HIPOT', 'GRAV_HEMAT', 'GRAV_MELEN', 'GRAV_METRO', 
        'GRAV_SANG', 'GRAV_AST', 'GRAV_MIOC', 'GRAV_CONSC', 'GRAV_ORGAO', 'DT_GRAV'
    ]

    for field in fields:
        # Remover linhas onde o campo √© obrigat√≥rio mas est√° nulo
        df = df[~(mandatory_conditions & df[field].isna())]
        # Remover linhas onde o campo est√° preenchido mas a condi√ß√£o n√£o √© satisfeita
        df = df[~((~mandatory_conditions) & df[field].notna())]

    return df
```

```python
# 35 - Exame sorol√≥gico (IgM) Chikungunya Soro 1 Data da Coleta 
# DT_CHIK_S1: Campo Essencial
# DT_CHIK_S1 >= DT_ALRM
# Habilitado apenas se campo ‚ÄòID_AGRAVO‚Äô = 2

df_grav['DT_CHIK_S1'] = pd.to_datetime(df_grav['DT_CHIK_S1'], errors='coerce')
df_grav['DT_ALRM'] = pd.to_datetime(df_grav['DT_ALRM'], errors='coerce')

# Condi√ß√µes para os campos obrigat√≥rios
chik_condition = (df_grav['ID_AGRAVO'] == 2)

# Remover linhas onde DT_CHIK_S1 √© nulo ou menor que DT_ALRM quando ID_AGRAVO √© 2

# Remover linhas onde DT_CHIK_S1 √© nulo ou menor que DT_ALRM quando ID_AGRAVO √© 2
df_grav = df_grav[~((chik_condition & df_grav['DT_CHIK_S1'].isna()) | (chik_condition & (df_grav['DT_CHIK_S1'] < df_grav['DT_ALRM'])))]

# Remover linhas onde DT_CHIK_S1 est√° preenchida mas ID_AGRAVO √© diferente de 2
df_chiks1 = df_grav[~((~chik_condition) & df_grav['DT_CHIK_S1'].notna())]
```

```python
# 36 - Exame sorol√≥gico (IgM) Chikungunya Soro 2 Data da Coleta 
# DT_CHIK_S2: Campo Essencial 
# DT_CHIK_S2 >= DT_CHIK_S1 
# Habilitado apenas se campo ‚ÄòID_AGRAVO‚Äô = 2


df_chiks1['DT_CHIK_S2'] = pd.to_datetime(df_chiks1['DT_CHIK_S2'], errors='coerce')

# Condi√ß√µes para os campos obrigat√≥rios
chik_condition = (df_chiks1['ID_AGRAVO'] == 2)

# Remover linhas onde DT_CHIK_S2 √© nulo ou menor que DT_CHIK_S1 quando ID_AGRAVO √© 2
df_chiks1 = df_chiks1[~((chik_condition & df_chiks1['DT_CHIK_S2'].isna()) | (chik_condition & (df_chiks1['DT_CHIK_S2'] < df_chiks1['DT_CHIK_S1'])))]

# Remover linhas onde DT_CHIK_S2 est√° preenchida mas ID_AGRAVO √© diferente de 2
df_chiks2 = df_chiks1[~((~chik_condition) & df_chiks1['DT_CHIK_S2'].notna())]
```

```python
def data_unabled(df):
    # Converter colunas de data para o formato datetime
    df['DT_PRNT'] = pd.to_datetime(df['DT_PRNT'], errors='coerce')

    # Condi√ß√µes para os campos obrigat√≥rios
    chik_condition = (df['ID_AGRAVO'] == 2)

    # Lista de campos a serem verificados
    fields_to_check = ['DT_PRNT', 'RES_CHIKS1', 'RES_CHIKS2', 'RESUL_PRNT']

    for field in fields_to_check:
        # Remover linhas onde o campo √© obrigat√≥rio mas est√° nulo
        df = df[~(chik_condition & df[field].isna())]
        # Remover linhas onde o campo est√° preenchido mas a condi√ß√£o n√£o √© satisfeita
        df = df[~((~chik_condition) & df[field].notna())]

    return df

df_unabled = data_unabled(df_chiks2)
```

```python
def idagravo_abled(df):

    # Condi√ß√µes para os campos obrigat√≥rios
    chik_condition = (df['ID_AGRAVO'] == 2)
    dengue_condition = (df['ID_AGRAVO'] == 1)

    # Lista de campos a serem verificados e suas condi√ß√µes
    fields_conditions = [
        ('DT_PRNT', chik_condition),
        ('RES_CHIKS1', chik_condition),
        ('RES_CHIKS2', chik_condition),
        ('RESUL_PRNT', chik_condition),
        ('DT_SORO', dengue_condition & (df['DT_SORO'] >= df['DT_ALRM'])),
        ('RESUL_SORO', dengue_condition),
        ('DT_NS1', dengue_condition & (df['DT_NS1'] >= df['DT_ALRM'])),
        ('RESUL_NS1', dengue_condition),
        ('HISTOPA_N', dengue_condition),
        ('IMUNOH_N', dengue_condition)
    ]

    for field, condition in fields_conditions:
        # Remover linhas onde o campo √© obrigat√≥rio mas est√° nulo
        df = df[~(condition & df[field].isna())]
        # Remover linhas onde o campo est√° preenchido mas a condi√ß√£o n√£o √© satisfeita
        df = df[~((~condition) & df[field].notna())]

    return df

df_agravo = idagravo_abled(df_unabled)
```

```python
def data_pcr(df):
    # Converter colunas de data para o formato datetime
    df['DT_PRNT'] = pd.to_datetime(df['DT_PRNT'], errors='coerce')
    df['DT_SORO'] = pd.to_datetime(df['DT_SORO'], errors='coerce')
    df['DT_NS1'] = pd.to_datetime(df['DT_NS1'], errors='coerce')
    df['DT_ALRM'] = pd.to_datetime(df['DT_ALRM'], errors='coerce')
    df['DT_VIRAL'] = pd.to_datetime(df['DT_VIRAL'], errors='coerce')
    df['DT_PCR'] = pd.to_datetime(df['DT_PCR'], errors='coerce')

    # Condi√ß√µes para os campos obrigat√≥rios
    chik_condition = (df['ID_AGRAVO'] == 2)
    dengue_condition = (df['ID_AGRAVO'] == 1)
    viral_pcr_condition = (df['ID_AGRAVO'].isin([1, 2]))

    # Lista de campos a serem verificados e suas condi√ß√µes
    fields_conditions = [
        ('DT_PRNT', chik_condition),
        ('RES_CHIKS1', chik_condition),
        ('RES_CHIKS2', chik_condition),
        ('RESUL_PRNT', chik_condition),
        ('DT_SORO', dengue_condition & (df['DT_SORO'] >= df['DT_ALRM'])),
        ('RESUL_SORO', dengue_condition),
        ('DT_NS1', dengue_condition & (df['DT_NS1'] >= df['DT_ALRM'])),
        ('RESUL_NS1', dengue_condition),
        ('HISTOPA_N', dengue_condition),
        ('IMUNOH_N', dengue_condition),
        ('DT_VIRAL', viral_pcr_condition & (df['DT_VIRAL'] >= df['DT_ALRM'])),
        ('RESUL_VI_N', viral_pcr_condition),
        ('DT_PCR', viral_pcr_condition & (df['DT_PCR'] >= df['DT_ALRM'])),
        ('RESUL_PCR_', viral_pcr_condition)
    ]

    for field, condition in fields_conditions:
        # Remover linhas onde o campo √© obrigat√≥rio mas est√° nulo
        df = df[~(condition & df[field].isna())]
        # Remover linhas onde o campo est√° preenchido mas a condi√ß√£o n√£o √© satisfeita
        df = df[~((~condition) & df[field].notna())]

    return df

# Aplicar a fun√ß√£o de valida√ß√£o e limpeza
df_pcr = data_pcr(df_agravo)
```
